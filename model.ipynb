```python
import os
import pandas as pd
from google import genai
from google.genai.errors import APIError

# YOUR_API_KEY = 

try:
    if not YOUR_API_KEY or YOUR_API_KEY == "YOUR_API_KEY":
        print("❌ Error: Harap masukkan API Key lo di variabel YOUR_API_KEY.")
        exit()

    client = genai.Client(api_key=YOUR_API_KEY)
    
    print("✅ Client berhasil diinisiasi menggunakan Direct API Key.")
    print("-" * 50)
    
except Exception as e:
    print(f"❌ Error inisiasi client. Cek lagi apakah API Key lo valid. Detail: {e}")
    exit()

try:
    models = client.models.list()
    
    available_models = []
    
    for model in models:
        if ('gemini-2.5' in model.name or 'imagen-3.0' in model.name):
            supported_methods = getattr(model, 'supported_generation_methods', getattr(model, 'supported_methods', []))
            
            supports_generate = 'generateContent' in supported_methods
            
            display_name = model.display_name if model.display_name else "N/A"
            
            available_models.append({
                "Model Name": model.name,
                "Display Name": display_name,
                "Supports generateContent": "✅" if supports_generate else "❌",
                "Input/Output Methods": supported_methods
            })
    
    # 4. Tampilkan hasilnya
    print("Daftar Model yang Tersedia (Standard Tier):")
    print("=" * 100)
    
    if available_models:
        try:
            df = pd.DataFrame(available_models)
            print(df.to_markdown(index=False)) 
        except ImportError:
            print("Instal Pandas (`pip install pandas`) untuk tampilan tabel yang lebih rapi.")
            for m in available_models:
                print(f"- {m['Display Name']} ({m['Model Name']}): Support Generate = {m['Supports generateContent']}")
    else:
        print("Tidak ada model 'gemini-2.5' atau 'imagen-3.0' yang ditemukan.")


except APIError as e:
    print(f"❌ Terjadi kesalahan API. Cek status API Key atau kuota lo. Detail: {e}")
except Exception as e:
    print(f"❌ Terjadi kesalahan tak terduga: {e}")
```
